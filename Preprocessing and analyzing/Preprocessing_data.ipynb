```python
import pandas as pd
# Chuy·ªÉn ƒë·ªïi, l·∫•y s·ªë t·ª´ c·ªôt l∆∞∆°ng d·∫°ng text ƒë·ªÉ chuy·ªÉn v·ªÅ d·∫°ng s·ªë
def convert_to_number(salary_str):
    multiplier = 1
    if 'tri·ªáu' and ',' in salary_str:
        multiplier = 10**5
    elif 'tri·ªáu' in salary_str:
        multiplier = 10**6
    elif 'ngh√¨n' in salary_str:
        multiplier = 10**3

    number_str = ''.join(filter(str.isdigit, salary_str))

    if not number_str:
        return 'Th·ªèa thu·∫≠n'

    number = int(number_str)
    return number * multiplier

data['Salary'] = data['priceNormal'].apply(convert_to_number)

data['Salary'].value_counts()
```

```python
import numpy as np
def convert_salary(salary):
    salary = salary.replace('.', '').replace('ƒë/th√°ng', '').replace('ƒë/ng√†y', '').strip()
    if 'Th∆∞∆°ng l∆∞·ª£ng' in salary:
      return 'Th·ªèa thu·∫≠n'
    if '-' in salary:
        # X·ª≠ l√Ω gi√° tr·ªã '6.000.000 - 15.000.000', ...
        salary_values = [float(value) for value in salary.split('-')]
        return int(sum(salary_values) / len(salary_values))
    elif 'tri·ªáu' in salary:
        # N·∫øu kh√¥ng ph·∫£i c√°c tr∆∞·ªùng h·ª£p tr√™n, gi·ªØ nguy√™n gi√° tr·ªã
        return np.nan
```

```python
doi_tac['Salary'] = doi_tac['priceNormal'].apply(convert_salary)
```

```python
def convert_to_number(salary_str):
    if isinstance(salary_str, float):
        return salary_str
    multiplier = 1
    if 'tri·ªáu' in salary_str and ',' in salary_str:
        multiplier = 10**5
    elif 'tri·ªáu' in salary_str:
        multiplier = 10**6
    elif 'ngh√¨n' in salary_str:
        multiplier = 10**3

    number_str = ''.join(filter(str.isdigit, salary_str))

    if not number_str:
        return 'Th·ªèa thu·∫≠n'

    number = int(number_str)
    return number * multiplier
doi_tac['Salary'] = doi_tac.apply(lambda row: convert_to_number(row['priceNormal']) if pd.isna(row['Salary']) else row['Salary'], axis=1)
```

```python
doi_tac.drop(columns = ['Unnamed: 0','Unnamed: 0.1','link', 'title', 'companyName', 'benefit','priceNormal'], inplace = True)
doi_tac['skill'] = 'Kh√¥ng'
len(doi_tac)
```
