```python
import pandas as pd
# Chuyển đổi, lấy số từ cột lương dạng text để chuyển về dạng số
def convert_to_number(salary_str):
    multiplier = 1
    if 'triệu' and ',' in salary_str:
        multiplier = 10**5
    elif 'triệu' in salary_str:
        multiplier = 10**6
    elif 'nghìn' in salary_str:
        multiplier = 10**3

    number_str = ''.join(filter(str.isdigit, salary_str))

    if not number_str:
        return 'Thỏa thuận'

    number = int(number_str)
    return number * multiplier

data['Salary'] = data['priceNormal'].apply(convert_to_number)

data['Salary'].value_counts()
```

```python
import pandas as pd
doi_tac = pd.read_csv("/content/data_cho_tot_doi_tac.csv")
```

```python
#Bỏ duplicate
doi_tac.drop_duplicates(inplace = True)
```

```python
import numpy as np
def convert_salary(salary):
    salary = salary.replace('.', '').replace('đ/tháng', '').replace('đ/ngày', '').strip()
    if 'Thương lượng' in salary:
      return 'Thỏa thuận'
    if '-' in salary:
        # Xử lý giá trị '6.000.000 - 15.000.000', ...
        salary_values = [float(value) for value in salary.split('-')]
        return int(sum(salary_values) / len(salary_values))
    elif 'triệu' in salary:
        # Nếu không phải các trường hợp trên, giữ nguyên giá trị
        return np.nan
```

```python
doi_tac['Salary'] = doi_tac['priceNormal'].apply(convert_salary)
```

```python
doi_tac['Salary'].unique()
```

```python
doi_tac['priceNormal'].unique()
```

```python
def convert_to_number(salary_str):
    if isinstance(salary_str, float):
        return salary_str
    multiplier = 1
    if 'triệu' in salary_str and ',' in salary_str:
        multiplier = 10**5
    elif 'triệu' in salary_str:
        multiplier = 10**6
    elif 'nghìn' in salary_str:
        multiplier = 10**3

    number_str = ''.join(filter(str.isdigit, salary_str))

    if not number_str:
        return 'Thỏa thuận'

    number = int(number_str)
    return number * multiplier
doi_tac['Salary'] = doi_tac.apply(lambda row: convert_to_number(row['priceNormal']) if pd.isna(row['Salary']) else row['Salary'], axis=1)
```

```python
doi_tac.drop(columns = ['Unnamed: 0','Unnamed: 0.1','link', 'title', 'companyName', 'benefit','priceNormal'], inplace = True)
doi_tac['skill'] = 'Không'
len(doi_tac)
```

```python
check_doi_tac = pd.read_csv("/content/data_luong_clean.csv")
check_doi_tac.drop_duplicates(inplace = True)
```

```python
#Merge 2 data đối tác và không đối tác
merged_df = pd.concat([doi_tac, check_doi_tac], ignore_index=True)
```
